# Radiação térmica: o começo de toda a física quântica


## O Contexto

<p style='text-align: justify;'>
O cenário da física no inicio do século XX era fantástico. Os físicos da época  estavam satisfeitos com toda a teoria classica sobre Mecânica e eletromagnetismo. O modelo discreto de niveis de energia de um atomo proposto por Boltzmann satisfazia as equações de Maxwell e havia um grande consenso sobre a natureza da luz: é onda! A física estatística ja havia avançado o suficiente para existir uma teoria muito concreta sobre a relação de temperatura com a energia cinética de gases. Existiam apenas dois problemas que seriam necéssarios para se ter todo o conhecimento sobre a natureza: O espectro da radiação de um corpo negro e o efeito fotoelétrico.
</p>
<p style='text-align: justify;'>
O século XIX foi muito importante para o desenvolvimento da termodinâmica clássica. Muita atenção era dada para os estudos das propriedades do calor em varios objetos. Nesse contexto, foi desenvolvido um modelo de corpo negro ideal: um objeto que absorve toda a radiação que o atinge e ao mesmo tempo que emite energia.Esse modelo propunha que a energia irradiada poderia ser escrita como ondas estacionárias dentro de uma cavidade, equivalente a um corpo negro. A energia era emitida na forma de ondas de frequência $\nu$, e deveria ser proporcional ao numero de modos de vibração dessa onda. A física estatística da época tinha como estabelecido que a energia media de cada modo é proporcional a kT. Mas os modos de vibração são proporcionais a frequência emitida, e por isso era de se esperar que as energias fossem cada vez mais altas a medida que as frequências fossem maiores. Isso não é o que acontece na realidade, até porque se um corpo negro emite um espectro continuo de frequências então esse modelo levaria a uma emissão de ultravioleta que não condiz com a realidade. 
</p>
<p style='text-align: justify;'>
Varios experimentos sobre o assunto foram realizados na época. Entre eles o de Balfour Stewart, que em 1858, experimentou a radiação termmica emitida e absorvida em placas polidas de diferentes substancias e as comparou com suberficies escuras de mesma temperatura. Houve também o experimento de Gustav Kirchhoff em 1959, que não sabendo do trabalho de Stewart, relatou a coincidência dos comprimentos de onda das linhas espectralmente resolvidas de absorção e de emissão de luz visível. 
</p>
<p style='text-align: justify;'>
Até 1900, quando Plank apresentou uma solução para esse problema, ninguém sabia explicar o porque do modelo téorico da época não condizia com o observado. Acontece que esse problema foi um grande marco para física quântica pois sua explicação leva a toda uma teoria que revolucionou toda a maneira como a natureza é entendida pela ciência.
</p>








## A Catastrofe
<p style='text-align: justify;'>
Um dos resultados mais brilhantes das equações de Maxwell é a previsão de ondas eletromagnéticas. Usando apenas as equações é possivel deduzir que a variação no campo elétrico de uma onda é dada por:
.
$$
\frac{\partial^2 E}{\partial x^2}+\frac{\partial^2 E}{\partial y^2}+\frac{\partial^2 E}{\partial z^2} =\frac{1}{c^2} \frac{\partial^2 s}{\partial t^2}
$$

Para o modelo em que um corpo negro é descrito como uma cavidade a solução da equação de onda deve ter amplitude 0 nas paredes da cavidade, até porque um valor diferente de zero implicaria na dissipação de energia, o que vai contra a suposição que o sistema esta em equilíbrio. Para formar uma onda estacionária, o caminho de reflexão ao redor da cavidade deve produzir um caminho fechado. Através dessas condições de contorno podemos escrever uma solução na forma:

$$
E=E_o sen \frac{n_1 \pi x}{L}  sen \frac{n_2 \pi y}{L}  sen \frac{n_3 \pi z}{L}  sen \frac{2 \pi c t}{\lambda} 
$$

E ao comparar essa solução com a equação de onda temos que:
 
$$
\begin{equation}
n_1^2+ n_2^2+ n_3^2=\frac{4L^2}{\lambda^2}
\label{n123} \tag{1}
\end{equation}
$$

Como descrito na seção anterior desse texto, é importente avaliar o número de modos possiveis para essa condição, isto é, contar todas as combinações possiveis para valores inteiros de n. Uma aproximação pode ser feita tratando o numero de combinações possiveis como o volume tridimencional de valores de n. Ao usar a relação de volume para uma esfera, com os valores de n como 3 eixos diferentes de um espaço tridimensional, é possivel escrever que:

$$
n=\frac{4 \pi}{3} (n_1^2+n_2^2+n_3^2)^{\frac{3}{2}}
$$

Porém, é necessario fazer algumas correções a respeito desse valor. Primeiro que um espaço tridimensional admite valores negativos para $n_1$, $n_2$ e $n_3$, o que na realidade não acontece e por isso consideramos apenas o espaço em que todos os n são positivos ($\frac{1}{8}$ do espaço total). Além disso, a onda pode estar polarizada em duas direções perpendiculares, e por isso o valor deve ser dobrado. O valor de n real fica:

$$
n=\frac{\pi}{3} (n_1^2+n_2^2+n_3^2)^{\frac{3}{2}}
$$

E de acordo com a equação  \ref{n123}:

$$
n=\frac{8 \pi L^3}{3 \lambda^3}
$$

Tendo desenvolvida uma expressão para o numero de ondas estacionarias em uma cavidade, é interessante conhecer a distribuição de numero de modos de acordo com o comprimento de onda. Para isso basta que:
$$
\begin{equation}
\frac{dn}{d\lambda}=\frac{d}{d\lambda}\frac{8 \pi L^3}{3 \lambda^3} = -\frac{8 \pi L^3}{ \lambda^4}
\label{dist} \tag{2}
\end{equation}
$$

E para obter a distribuição por unidade de volume é suficiente apenas dividir a equação \ref{dist} pelo volume da cavidade (a qual consideraremos cúbica, mas que no final não faz diferença):
$$
\frac{1}{L^3}(-\frac{8 \pi L^3}{ \lambda^4})=-\frac{1}{L^3}\frac{dN}{d\lambda}=\frac{8\pi}{\lambda^4}
$$

Aqui, é necessario introduzir o teorema da equipartição. Em 1845 foi desenvolvido por John Wum teorema que relaciona a temperatura de um sistema com a energia media de um modo de vibração atraves da constante de Boltzmann.
Como a cavidade possui n modos de vibração, $\langle E_{TOTAL} \rangle=nkT$. Assim, colocando u como a densidade de energia:

$$
\frac{du}{d\lambda}=\frac{1}{L^3}\frac{dE}{d\lambda}=-kT\frac{1}{L^3}\frac{dN}{d\lambda} = \frac{8 \pi kT}{\lambda^4}
$$

Essa expressão tambem pode ser escrita em função da frequência apenas usando a regra da cadeia:
$$
\frac{du}{d\nu}=\frac{du}{d\lambda}\frac{d\lambda}{d \nu}=\frac{du}{d\lambda}\frac{d(c/\nu)}{d \nu}=\frac{du}{d\lambda}\frac{c}{\nu^2}=\frac{8 \pi kT\nu^4}{c^4}\frac{c}{\nu^2}=kT\frac{8 \pi \nu^2}{c^3}
$$

Se considerar que a energia irradiada perpendicular a um pequeno incremento de area, é possivel perceber que metade da densidade de energia está indo contra as paredes enquanto metade está saindo do sistema em equilibrio termico. É interessante observar a potência a partir de um ponto fora da cavidade. Se esse ponto está a uma distancia x da da area de emissão, A potência irradiada por unidade de comprimento de onda é igual a metade da densidade de energia por comprimento de onda vezes a razão entre o volume e o tempo para a energia irradiada chegar ao ponto x. Assim:
$$
R(\lambda)=\frac{1}{2}\frac{du}{d\lambda}\frac{Ax}{t}=\frac{1}{2}\frac{du}{d\lambda}\frac{Act}{t}=\frac{du}{d\lambda}\frac{Ac}{2}
$$
Fazendo uma correção angular em relação a area efetiva em um angulo $\theta$:
$$
R(\lambda)=\frac{du}{d\lambda}\frac{Ac}{2}cos^2\theta
$$
Fazendo a media sobre todos os angulos:
$$
R(\lambda)=\frac {\frac{du}{d\lambda}\frac{Ac}{2}\langle cos^2\theta \rangle}{Area}=\frac{du}{d\lambda} \frac{c}{4}=\frac{8 \pi kT}{\lambda^4} \frac{c}{4} = \frac{2 \pi c k T}{\lambda^4}
$$

Essa equação é a equação de Rayleigh-Jeans. Como é possivel perceber pela dedução, ela deriva do eletromagnetismo clássico e algumas propriedades matématicas. É possivel perceber também que quanto menor o comprimento de onda maior a radiancia, e quando o comprimento de onda tende a zero a radiancia tente ao infinito! Com um código simples em python é possivel observar o crescimento absurdo da radiancia a medida que o comprimento de onda fica menor.
</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.colors
import os
cmap = plt.cm.rainbow
%matplotlib inline
```

```python
c=3*10^8
k=1.38064852*10**-23
def R(l,T):
    return 2*np.pi*c*T/l**4

L=[10**i for i in np.linspace(-5,5,100)]
T=[1000*i for i in range(1,7)]
Rad=[]
C=[]
for t in T:
    Rt=[]
    ct=[]
    for l in L:
        Rt.append(np.log(R(l,t))/np.log(10))
        ct.append(t*50)
    Rad.append(Rt)
    C.append(ct)

for i in range(len(T)):
    plt.plot(L,Rad[i], label='T='+str(T[i]))
plt.grid()
plt.xlabel('Comprimento de onda (m)')
plt.ylabel('logaritmo da Radiância')


plt.legend()
```

<p style='text-align: justify;'>
Como é possivel ver no grafico anterior, para valores muito pequenos de comprimento de onda a radiancia estrapola para valores absurdos! Tal problema é inexplicavel apenas com a física clássica e ficou conhecido como catástrofe do ultravioleta. 
</p>

## A Solução
<p style='text-align: justify;'>
A solução para o problema do corpo negro foi proposta por Plank numa reunião da sociedade alemã de física no final de 1900. Com uma proposta radical e sem uma explicação concreta de sua motivação, Plank propôs que a energia emitida e absorvida por um corpo negro poderia ser quantizada, como em pequenos pacotes de energia. Ele propôs que a energia deveria desses pacotes deveria ser proporcional a frequência da onda sendo emitida. Essa hipótese também implica que para uma certa temperatura existe uma intensidade máxima de radiação emitida pelo corpo negro. 
</p>
<p style='text-align: justify;'>
A medida que a frequência eletromagnetica da radiação cresce, a magnitude associada com a energia também cresce. Ao assumir que a energia pode ser emitida apenas em multiplos da frequência, Plank deduziu uma equação que encaixava muito bem com a curva experimental de experimentos de corpo negro. A lógica é a seguinte: em baixas temperaturas, a radiação é emitida apenas com frequencias baixas, o que corresponde a pacotes de baixa energia. A medida que a temperatura do objeto cresce, existe também um crescimento na probabilidade do corpo negro emitir um pacote de alta energia (i.e. um pacote de energia de alta frequência). E em qualquer tempratura é mais provavel um corpo negro emitir uma quantidade maior de pacotes de baixa energia do que uma quantidade menor de pacotes de alta energia. Nesse sentido, existe também uma frequência mais provavel de emissão, que é a frequência para qual se tem um pico na distribuição espectral. Esse pico tende a aumentar de frequência a medida que a temperatura aumenta.
</p>
<p style='text-align: justify;'>
A hipotese de plank explicava a distribuição expectral como uma consequência da oscilação de eletrons. Max Plank se concentrou em criar um modelo para cargas oscilantes que deveria existir nas paredes da cavidade de um corpo negro, descrito na seção anterior. Assim, os eletrons irradiam calor para dentro e - em equilíbrio termodinâmico - emitem radiação para fora. Nesse sentido, os eletrons não irradiariam energia de forma continua como era o esperado pela teoria clássica, mas em intervalos quantizados de energia (chamados de quanta). Esses intervalos de energia possuiam a mesma magnitude dos pequenos pacotes de energia que Plank propos para a frequência emitida pelo corpo negro, $E=h\nu$.
</p>
<p style='text-align: justify;'>
A parte matemática é um pouco mais complexa que a da seção anterior. Porém, é possivel deduzir a expressão de Plank para densidade de energia a partir da lei de distribuição de boltzmann:

$$
P(E)=\frac{e^{\frac{-E}{kT}}}{kT}
$$

O teorema da equipartição citado anteriormente e os intervalos de energia discretos $\Delta E=h\nu$. Plank descreveu a radiância espectral:

$$
\rho(\nu)==\frac{2 h  \nu^3}{c^2}\frac{1}{e^{\frac{h\nu}{kT}}-1}
$$

Também é posível escrever a densidade de energia emfunção do comprimento de onda. Fica:

$$
u(\lambda)=\frac{2 c^2 h}{\lambda^5}\frac{1}{e^{\frac{hc}{\lambda kT}}-1}
$$

Usando mais um código em python é possivel observar a coerencia da Lei de Plank para radiância espectral.
</p>

```python
h=6.63*10**-34
def u(wav, T):
    a = 2.0*h*c**2
    b = (h*c/(wav*k*T))
    I = a/ ( (wav**5) * (np.exp(b) - 1.0) )
    return I


L=np.arange(1e-9, 3e-6, 1e-9)
T=[1000*i for i in range(1,7)]
Rad=[]
for Tt in T:
    Rad.append(u(L,Tt))
for i in range(len(T)):
    plt.plot(L*1e9,Rad[i], label='T='+str(T[i])+'K')
    
plt.legend()
plt.grid()
plt.xlabel('Comprimento de onda (m x 10^-9)')
plt.ylabel('densidade de energia')
plt.legend()
```


<p style='text-align: justify;'>
Como é evidente no gáfico anterior a hipótese de Plank faz muito mais sentido que a equação de Rayleigh-Jeans. É possivel notar também que o pico da curva se direciona para a esquerda a medida que a temperatura aumenta, isto é, quanto maior a temperatura maior a probabilidade de um corpo negro emitir uma onda de frequência maior. Outra coisa que é possivelperceber é que a maior parte dos comprimentos de onda emitidos por um corpo negro de temperatura 5000K é entre 400 nm e 800nm que é a faixa em que a luz visivel é emitida, e que corpo negro emite nessa faixa de temperatura? O sol! Estrelas no geral não são corpos negros perfeitos, mas podemos trata-las como tal para obter suas temperaturas. É por causa dessa lei inclusive que estrelas azuis são muito mais quentes que estrelas vermelhas, a a frequência de maior emissão tende para o violeta a medida que a temperatura aumenta.
</p>
<p style='text-align: justify;'>
Essa  hipótese não foi muito bem aceita pelos físicos da época pois a natureza discreta é contraintuitiva em relação a física da época. No entanto, a formulação de PLank acabou sendo provisoriamente tolerada pelos físicos, pois era
a única de que se dispunha para dar conta dos fatos. O cenario muda alguns anos depois, quando Einstein usa essa mesma hipótese para explicar o outro grande problema da física na época: o efeito fotoelétrico.
</p>


## Confirmações
<p style='text-align: justify;'>
A Lei de Plank para distribuição de Energia não somente é condizente com o espectro de emissão de um corpo negro como também com a lei de Stefan e com a lei de deslocamento de Wien. Essas duas ultimas são resultados empiricos que são muito usados na astronomia para calcular temperatura de estrelas.
 A lei de Stefan relaciona a Potencia emitida por area com a Temperatura do corpo. Assim, para deduzir a lei de Stefan, basta integrar a função obtida por Plank (que descreve as densidades de energia para um comprimento de onda ou frequencia) entre 0 e infinito e multiplicar por c/4. Fica:
$$
R_T=\frac{c}{4}\rho_T=\frac{c}{4}= \int_{0}^{\infty} \rho(\nu)d\nu=\frac{c}{4}\frac{8 h \pi }{c^3}\int_{0}^{\infty}\frac{\nu^3}{e^{\frac{h\nu}{kT}}-1}d\nu\\
\alpha=\frac{h\nu}{kT}\\
R_T=\frac{c}{4}\frac{8 h \pi }{c^3}(\frac{kT}{h})^4\int_{0}^{\infty}\frac{\alpha^3}{e^{\alpha}-1}d\alpha\\
\int_{0}^{\infty}\frac{\alpha^3}{e^{\alpha}-1}d\alpha=\frac{\pi^4}{15}\\
R_T=\frac{c}{4}\frac{8 h \pi }{c^3}(\frac{kT}{h})^4\frac{\pi^4}{15}=\frac{2 \pi^5 k^4}{15 c^2 h^3}T^4=5,67 \cdot 10^{-8}T^4
$$
Isso prova que a lei de Plank confirma a lei de Stefan.  Já a lei de deslocamento de Wien descreve a frequência ou o compriento de onda que é mais emitido por um corpo negro, e para deduzi-la basta derivar a formula de Plank em relação a frequência ou comprimento de onda e igualar a zero. Assim:
$$
\frac{d}{d\nu}\rho(\lambda)=0\\
3(e^{\frac{h\nu}{kT}}-1)-\nu\frac{h}{kT}e^{\frac{h\nu}{kT}}=0\\
\alpha=\frac{h\nu}{kT}\\
3(e^\alpha-1)-\alpha e^\alpha =0\\
\frac{\alpha}{3}+e^{-\alpha}=1
$$

Essa equação é uma equação trancendental, isto é, não possui solução analitica. Mas ela possui uma solução numérica em que $\alpha=4,965$. A equação então fica:
$$
\alpha=4,965=\frac{h\nu}{kT}\\
\nu=4,965\frac{k}{h}T\\
\lambda=\frac{ch}{4,965k}\frac{1}{T}\\
$$
É possivel usar a lei de Plank confirmar tambem a própria relação de Rayleigh-Jeans, até porque para valores muito altos de comprimento de onda (ou baixos de frequência) a equação de Rayleigh-Jeans faz sentido. Para tal é só usar a serie de taylor para a função exponencial, como os valores ficam muito pequenos da para truncar o polinomio no segundo termo. Fica:
$$
R(\lambda)=\frac{2 h c^2 \pi}{\lambda^5}\frac{1}{e^{\frac{hc}{\lambda kT}}-1}=\frac{2 c^2 h\pi}{\lambda^5}\frac{1}{1+\frac{hc}{\lambda kT}-1}=\frac{2 c^2 h\pi}{\lambda^5}\frac{1}{\frac{hc}{\lambda kT}}=\frac{2 \pi kcT}{\lambda^4}\\
$$
</p>

## Aplicações
<p style='text-align: justify;'>
Além de ser um estopim da mecânica quântica e por consequência ser uma das causas de toda a evolução científica do século passado, a hipótese de Plank e sua equação têm aplicações diretas na Biologia, Astrofísica e na Cosmologia.
O universo é o experimento de corpo negro que se encaixa mais perfeitamente com a curva de Plank que conhecemos hoje. Isso porque a terra está sendo sempre atingida pelo que é conhecido como radiação cosmica de fundo (CMB). Essa radiação é uma reliquia da formação do universo, se encontra na faixa de microondas e é através dela que é possivel estimarmos a Temperatura dos primordios do universo. Existem alguns satelites que são destinados a captar essa radiação e diferenciar ela de outras ondas eletromagnéticas que atingem a terra. Um deles é o Cosmic Background Explorer (COBE ou Explorer 66), um sátelite que operou entre 1989 e 1993 pela NASA com o objetivo de justamente observar a radiação cósmica de fundo.
Parte dos dados que foram captados nesse experimento estão disponiveis na internet e podem ser processados rapidamente atraves de um código em python. Os dados a seguir foram obtidos através do site https://aether.lbl.gov/www/projects/cobe, relacionam a intensidade luminosa($erg \cdot cm^{-2} \cdot(s \cdot sr \cdot )^{-1}$) com comprimento de onda(cm).
</p>

```python
wavl,Int = np.loadtxt('data.txt',unpack = True)
```

```python
plt.scatter(wavl,Int*10**3, s=5, c='r')
plt.grid()
plt.xlabel('Comprimento de onda (cm)')
plt.ylabel('Intensidade luminosa')
```

Através desses dados é possivel obter a temperatura do universo em seus primordios! Por exemplo: se interpolarmos a curva e e buscarmos o comprimento de onda para qual a intensidade é máxima da para obter, através da lei de Wien, uma estimativa de quanto deve ser essa temperatura. 

```python
from scipy import interpolate
f = interpolate.interp1d(wavl, Int)
xnew = np.linspace(min(wavl), max(wavl),1000)
ynew = f(xnew)

from scipy.constants import h 
from scipy.constants import c 
from scipy.constants import k 
def planck(lambdanm, T):
    intensity = 2.0*h*c**2/ ( (lambdanm**5) * (np.exp(h*c/(lambdanm*k*T)) - 1.0) )
    return intensity


plt.scatter(wavl*10**-2,Int*10**3, s=5, c='r')
plt.plot(xnew*10**-2,ynew*10**3)
plt.grid()
maxlen=2*xnew[np.argmax(ynew)]*10**-4
T=0.002898/maxlen
plt.xlabel('Comprimento de onda (cm)')
plt.ylabel('Intensidade luminosa')
print('T=',T)
```

<p style='text-align: justify;'>
A estimativa calculada foi de 2.56 Kelvin, ao passo que o valor estabelecido hoje é de 2.7 Kelvin (Mais detalhes sobre esse experimento estão disponiveis em https://ui.adsabs.harvard.edu/abs/1994ApJ...420..439M/abstract). Esses valores nos permitem também calcular a energia do universo e a temperatura para quando se expandir ao dobro do volume atual (Questão 9.4 e 9.5, EUF 2009/2). Ao integrar a equação de Plank para densidade de energia se obtem a seguinte relação:
$$
\frac{E}{V}=\frac{4 \sigma T^4}{c}\\
E=\frac{4 V\sigma T^4}{c}\\
$$
Se considerarmos o universo como uma casca esférica de raio $10^{28}cm$, o volume será $V=\frac{4}{3} \pi r^3$ e a Energia fica num valor próximo de $2,6 \cdot 10^{65} J$. Já pensando no universo como uma expansão adiabática, temos que:
$$
\delta Q=TdS=0 \Rightarrow dS=0
$$
Podemos considerar que o universo é um gás de fotons e que portanto a entropia fica proporcional ao volume e a temperatura elevada a terceira potência.
$$
S=CVT^3 \Rightarrow dS=CT^3 dV + 3CVT^2dT=0 \Rightarrow cT^3dV=-3CT^2VdT \Rightarrow \frac{dV}{V}=\frac{3dT}{T}\\
ln V=-3lnT+lnG \Rightarrow V=GT^{-3} \Rightarrow T=(\frac{G}{2V_o})^{\frac{1}{3}}=2^{\frac{-1}{3}} T_o \approx 2.14 K
$$

(Questão retirada de https://www.ifsc.usp.br/posgraduacao/images/stories/pdf/prova_euf_2009_2.pdf)

A relação de Plank sobre o espectro de emissão de um corpo negro também é muito util para os astronomos, tendo em vista que ela permite medir a temperatura de estrelas e outros objetos cósmicos, que estão a distâncias enormes da terra, simplesmente medindo a cor de sua luz, isto é, o comprimento de onda mais emitido. Saber a temperatura de estrelas, de gáses interestelares e outros objetos astronômicos foi crucial para o nosso entendimento sobre os proessos físicos desses espaços. https://arxiv.org/abs/1711.01122 nesse link contem um artigo em que dois astronomos mediram o espectro de algumas estrelas e observaram a alta coerência com o espectro de um corpo negro, medindo também sua temperatura.
